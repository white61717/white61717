```python
def knights(saying):
    def inner(quote):
        return "we are %s" % quote +'hh'
    return inner(saying)
```

```python
a = knights(87)
type(a)
```

```python
def knights2(saying):
    def inner2():
        return "we are %s" % saying +'hh'
    return inner2
```

```python
knights2(87)
```

```python
a = knights2(87)
type(a)
```

```python
def edit_story(words, func):
    for word in words:
        print(func(word))
```

```python
stairs = ['a','b','c','d']
```

```python
def enliven(word):
    return word.capitalize()
```

```python
edit_story(stairs, enliven)
```

```python
edit_story(stairs, lambda word:word.capitalize())
```

```python
for word in stairs:
    print(word.capitalize())
```

```python
def yield_fun(n,i):
    while n < i:
        yield n
        n+=1
```

```python
generator = yield_fun(3,8)
for x in generator:
    print(x)
```
