```python
import numpy as np
import pandas as pd
```

```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
print(pop.head())
print()
print(areas.head())
print()
print(abbrevs.head())
```

```python
#美國各州及地區根據2010人口密度排名
merged = pd.merge(pop, abbrevs, how='outer', left_on='state/region', right_on='abbreviation').drop('abbreviation', axis=1)
merged
```

```python
merged.isnull().any()
```

```python
merged[merged['state'].isnull()]
```

```python
abbrevs
```

```python
merged.loc[merged['state'].isnull(), 'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'USA','state'] = 'United State'
merged.loc[merged['state/region'] == 'PR','state'] = 'Puerto Rico'
```

```python
merged
```

```python
merged.isnull().any()
```

```python
merged2 = pd.merge(merged, areas, on='state', how='left')
merged2['人口密度'] = merged2['population'] / merged2['area (sq. mi)']
merged2
```

```python
merged2.loc[merged2['area (sq. mi)'].isnull(),'state']
```

```python
merged2['state'][merged2['area (sq. mi)'].isnull()]
```

```python
merged2.dropna(inplace=True)
merged2
```

```python
data2010 = merged2.loc[(merged2['ages'] == 'total') & (merged2['year'] == 2010),]
data2010
# merged2.query("ages == 'total' & year == 2010")
```

```python
data2010.set_index('state', inplace = True)
```

```python
data2010
```

```python
density = data2010['population'] / data2010['area (sq. mi)']
density
```

```python
density.sort_values(ascending=False, inplace = True)
density
```
