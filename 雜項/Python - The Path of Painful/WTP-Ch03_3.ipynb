```python
#p175
#樞紐分析表
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
age = 10 * (titanic['age'] // 10 )
age = age.astype(str) + 'up'
age.fillna('no data')
age.name = 'age'
age
```

```python
a = titanic.groupby(['sex', age])['survived'].count().unstack().fillna(0)
a /= a.values.sum() / 100
a
```

```python
titanic[(titanic['sex'] == 'male') & (titanic['age'] > 0) & (titanic['age'] < 10)]
```

```python
titanic.groupby('sex')['survived'].mean()
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
titanic['survived']
```

```python
titanic[['survived']]
```

```python
titanic.groupby(['sex', 'class'])['survived'].mean().unstack()
```

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.groupby(['sex', age, 'class'])['survived'].mean().unstack().fillna(0)
```

```python
titanic.pivot_table('survived', ['sex', age], 'class')
```

```python
titanic
```

```python
fare = pd.qcut(titanic['fare'], 2)
fare
```

```python
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

```python
titanic.groupby(['sex', age, fare, 'class'])['survived'].mean().unstack()
```

```python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived': sum, 'fare': 'mean'}) 
```

```python
titanic.groupby(['sex', 'class']).aggregate({'survived': sum, 'fare': 'mean'})
```

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
```
